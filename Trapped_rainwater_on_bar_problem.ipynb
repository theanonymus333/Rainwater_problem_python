### <span style = "color:red;">Problem Statement: </span>
### Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

### <span style = "color:red;">Input: </span> height  =  [0,1,0,2,1,0,1,3,2,1,2,1] </span>
### <span style = "color:red;">Output : </span>  6
### The below elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.

### As shown above, we can see that the plots are made on the basis of the height list which was given. Here, after we plotted this, let's assume that rain falls and fills the gaps and spaces found in between.
### We need to start with an algorithm, which can actually set our path easy for the code:

### <span style = "color:red;"> Algorithm: </span>
###  * Find maximum height of bar from the left end upto an index i in the array \text{left\_max}left_max.
###  * Find maximum height of bar from the right end upto an index i in the array \text{right\_max}right_max.
###  * Iterate over the \text{height}height array and update ans:
###  * Add \min(\text{left\_max}[i],\text{right\_max}[i]) - \text{height}[i]min(left_max[i],right_max[i])−height[i] to          \text{ans}ans


### <span style = "color:red;"> Code: </span>

```python
water_trapped = 0
height = [0,1,0,2,1,0,1,3,2,1,2,1]
for i in range(1,len(height)-1):
    curr_bar_height=height[i]
    max_left=max(height[0:i])
    max_right=max(height[i+1:len(height)])
    if (curr_bar_height<max_left and curr_bar_height<max_right):
        water_trapped+=min(max_left, max_right) - curr_bar_height
print(water_trapped)
```
